### Styling and Themes in Seaborn

```python
# Let us try applying these themes

#load dataset tips 
tips = sns.load_dataset('tips')

# preset themes
sns.set_style('darkgrid')

# plot 
sns.stripplot(x="day", y="total_bill", data=tips)
```

### Seabornâ€™s plotting functions
Relational plots

```python
#scatterplot() ( kind=â€scatterâ€)
#lineplot() ( kind=â€lineâ€)
# The default plot kind is a scatterplot:
sns.relplot(x="total_bill", y="tip", hue="day", data=tips,col="time")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

### lineplot (axes-level function)

```python
sns.lineplot(data=tips, x="total_bill", y="tip")
```

# Distribution plots
#Suitable for comparing range and distribution for groups of numerical data

#displot -Figure-level interface for drawing distribution plots onto a FacetGrid.
#histplot -Plot univariate or bivariate histograms to show distributions of datasets.
#ecdfplot - Plot empirical cumulative distribution functions.
#rugplot -Plot marginal distributions by drawing ticks along the x and y axes.

```python
# The default plot kind is a histogram:
penguins = sns.load_dataset("penguins")
sns.displot(data=penguins, x="flipper_length_mm")
```

```python
#ecdfplot (axes-level function)

sns.ecdfplot(data=penguins, x="flipper_length_mm")
```

```python
#kdeplot (axes-level function)
sns.kdeplot(data=penguins, x="flipper_length_mm")

```

```python
#Categorical plots
#catplot Figure-level interface for drawing categorical plots onto a FacetGrid.
#stripplot Draw a scatterplot where one variable is categorical.
#swarmploT -Draw a categorical scatterplot with non-overlapping points.
#boxplot -Draw a box plot to show distributions with respect to categories.
#violinplot - Draw a combination of boxplot and kernel density estimate.
#boxenplot -Draw an enhanced box plot for larger datasets.
#pointplot -Show point estimates and confidence intervals using scatter plot glyphs.
#barplot -Show point estimates and confidence intervals as rectangular bars.
#countplot -Show the counts of observations in each categorical bin using bars.
```

```python
# catplot default 
exercise = sns.load_dataset("exercise")
g = sns.catplot(x="time", y="pulse", hue="kind", data=exercise)
```

```python
#Boxplot (axes-level function)

sns.boxplot(x=tips["total_bill"])
```

```python
#Barplot ((axes-level function)

sns.barplot(x="day", y="total_bill", data=tips)
```

### Multi-plot grids

```python
# Calling the constructor requires a long-form data object. This initializes the grid, but doesnâ€™t plot anything on it:
sns.FacetGrid(tips)

#Assign column and/or row variables to add more subplots to the figure:
face=sns.FacetGrid(tips, col="time", row="sex")

#To draw a plot on every facet, pass a function and the name of one or more columns in the dataframe to FacetGrid.map()
face.map(sns.scatterplot, "total_bill", "tip")
```
